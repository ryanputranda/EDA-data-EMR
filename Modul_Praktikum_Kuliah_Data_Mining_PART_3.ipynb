# **Melakukan One Hot Encoding**

```python
from sklearn.preprocessing import LabelEncoder

dataolahencoding = data_test.copy()

encoder = LabelEncoder()
dataolahencoding['diagnosa_ENCODE'] = encoder.fit_transform(dataolahencoding['diagnosa'])
dataolahencoding['institusi_ENCODE'] = encoder.fit_transform(dataolahencoding['institusi'])
dataolahencoding['area_ENCODE'] = encoder.fit_transform(dataolahencoding['area'])
dataolahencoding['tipe_ENCODE'] = encoder.fit_transform(dataolahencoding['tipe'])
dataolahencoding['biaya_berobat_ENCODE'] = encoder.fit_transform(dataolahencoding['biaya_berobat'])
dataolahencoding['interpretasi_biaya_ENCODE'] = encoder.fit_transform(dataolahencoding['interpretasi_biaya'])
dataolahencoding['interpretasi_diagnosa_ENCODE'] = encoder.fit_transform(dataolahencoding['interpretasi_diagnosa'])
```

```python
dataolahencoding.head()
```

# **Memecah data menjadi 2, yaitu: atribut dan kelas target**

```python
x = dataolahencoding[['diagnosa_ENCODE','institusi_ENCODE', 'area_ENCODE', 'tipe_ENCODE', 'biaya_berobat_ENCODE']] #atribut
y = dataolahencoding[['interpretasi_biaya_ENCODE']] #kelas target 1
z = dataolahencoding[['interpretasi_diagnosa_ENCODE']]  #kelas target 2
```

# **Split train test**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# **Klasifikasi Naive Bayes Standart**

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(x_train, y_train.values.ravel())
y_pred = model.predict(x_test)
```

```python
import pandas as pd
import numpy as np

random_indices = np.random.choice(len(x_test), size=10, replace=False)
random_x_test = x_test.iloc[random_indices]
random_y_pred = model.predict(random_x_test)
random_y_test = y_test.iloc[random_indices]


decoded_y_pred = encoder.inverse_transform(random_y_pred)
decoded_y_test = encoder.inverse_transform(random_y_test.values.ravel())
results_df = pd.DataFrame({
    'Target Biaya': decoded_y_test,
    'Prediksi Biaya': decoded_y_pred
})

results_df
```
